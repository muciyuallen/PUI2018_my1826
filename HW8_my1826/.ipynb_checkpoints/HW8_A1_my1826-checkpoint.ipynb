```python
import os

def downloadCitiBike(datestring):
    #first check if the file is already there
    if not os.path.isfile(os.getenv("PUIDATA") + "/" + datestring + '-citibike-tripdata.csv'):
        #check if zip already exists
        if os.path.isfile(datestring + '-citibike-tripdata.csv.zip'):
            print("The zipfile already exisits.")
            os.system("unzip " + datestring + '-citibike-tripdata.csv.zip')
            os.system("mv " + datestring + "-citibike-tripdata.csv " + os.getenv("PUIDATA")+ "/" + datestring + '-citibike-tripdata.csv')
            #check if the file has been downloaded and unzipped
            df = pd.read_csv(os.getenv("PUIDATA")+ "/" + datestring + '-citibike-tripdata.csv')
            return df
        else:
            print("The CitiBike file is not there, now downloading...")
            os.system("curl -O https://s3.amazonaws.com/tripdata/" + datestring + "-citibike-tripdata.csv.zip")
            os.system("unzip " + datestring + '-citibike-tripdata.csv.zip')
            os.system("mv " + datestring + "-citibike-tripdata.csv " + os.getenv("PUIDATA")+ "/" + datestring + '-citibike-tripdata.csv')
            #check if the file has been downloaded and unzipped
            df = pd.read_csv(os.getenv("PUIDATA")+ "/" + datestring + '-citibike-tripdata.csv')
            return df
    else:
        #report that the file has already been there
        print('The file has already been there.')
        df = pd.read_csv(os.getenv("PUIDATA")+ "/" + datestring + '-citibike-tripdata.csv')
        return df

citibike = downloadCitiBike('201804')
citibike.head()
```

```python
def get_trip_counts(bike_data):
    # make a DataFrame with locations for each bike station
    locations = bike_data.groupby("start station id").first()
    locations = locations.loc[:, ["start station latitude",
                                 "start station longitude",
                                 "start station name"]]    
    # count trips for each destination
    departure_counts =  bike_data.groupby("start station id").count()
    departure_counts = departure_counts.iloc[:,[0]]
    departure_counts.columns= ["Departure Count"]
    
    # count trips for each origin
    arrival_counts =  bike_data.groupby("end station id").count().iloc[:,[0]]
    arrival_counts.columns= ["Arrival Count"]

    #join departure counts, arrival counts, and locations
    trip_counts = departure_counts.join(locations).join(arrival_counts)
    
    
    return trip_counts
```

```python
def plot_station_counts(trip_counts):
    # generate a new map
    folium_map = folium.Map(location=[40.7291, -73.9965],
                        zoom_start=13,
                        tiles="Stamen Toner")

    # for each row in the data, add a cicle marker
    for index, row in trip_counts.iterrows():
        # calculate net departures
        net_departures = (row["Departure Count"]-row["Arrival Count"])
        
        # generate the popup message that is shown on click.
        popup_text = "{}<br> total departures: {}<br> total arrivals: {}<br> net departures: {}"
        popup_text = popup_text.format(row["start station name"],
                          row["Arrival Count"],
                          row["Departure Count"],
                          net_departures)
        
        # radius of circles
        radius = net_departures/20
        
        # choose the color of the marker
        if net_departures>0:
            # color="#FFCE00" # orange
            # color="#007849" # green
            color="#E37222" # tangerine
        else:
            # color="#0375B4" # blue
            # color="#FFCE00" # yellow            
            color="#0A8A9F" # teal
        
        # add marker to the map
        folium.CircleMarker(location=(row["start station latitude"],
                                      row["start station longitude"]),
                            radius=radius,
                            color=color,
                            popup=popup_text,
                            fill=True).add_to(folium_map)
    return folium_map
```

```python
trip_counts = get_trip_counts(citibike)
plot_station_counts(trip_counts)
```

In the map, orange circles refer to the station that has more departure trips than arrival trips, blue vice versa. The radius of the circle indicates the magnitude of net departure. 
The analysis of the influx and outflow of citi bikes between docks have some real-world meaning. First of all, we can see that the popular arrival destination locate in midtown and upper west side. Bike management team can consider put fewer bicycles in these locations in the beginning of the day. Also, from the distribution of the blue and orange cycles, city planners can observe some popular bikes routes, such as from upper east to upper west or from Bushwick to Williamsburg. Then they could consider renovate or improve some of these bike routes for better transit safety and efficiency.
